# Crescimento Global de Data Center Hyperscale

```python
import matplotlib.pyplot as plt
import csv

a = []
b = []

with open('data/dc_cloud_traffic_growth.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=';')
    for row in plots:
        a.append(float(row[0]))
        b.append(float(row[1]))
        

plt.bar(a,b, label='30% de Carga 2015-2020')
plt.xlabel('Ano')
plt.ylabel('Número')

plt.title('Crescimento Global da Demanda no Data Center')
plt.legend()
plt.show()
```
